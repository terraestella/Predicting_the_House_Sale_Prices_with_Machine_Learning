## Predicting the House Sale Prices with Machine Learning (Linear Regression Model)

This project will try to predict the house sale prices using the dataset of marketed houses in Ames, Iowa. The dataset consist of 82 features for each house sold, including the price it was sold for.
The steps to be taken in order to predict the sale price of a new house are as follows:

1. Analyzing data and understanding the features describing the houses
2. Data processing and preparing for deeper analysis
3. Choosing the features to be used in predictive modelling
4. Building the model and evaluating results
5. Answering which properties of a house most affect the final sale price

**Dataset:**
- Sold houses in Ames, Iowa - 2930 rows, 82 columns
- Each row in the dataset describes the properties (the columns) of a single house, as well as the amount it was sold for

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading data
data = pd.read_csv('AmesHousing.txt', delimiter='\t')
pd.set_option('display.max_columns', None)
print(data.shape)
data.head()
```

```python
# splitting data (50:50)
train = data[:1460]
test = data[1460:]

print(train.shape)
print(test.shape)
```

```python
data.info()
```

[data dictionary](https://s3.amazonaws.com/dq-content/307/data_description.txt)

#### Analyzing Data Correlations

- using corr() function to analyze correlations among all of the features (columns)
- sorting values in descending order to understand what features correlate the most with the selling price of a house

```python
train_corr = train.corr()['SalePrice'].sort_values(ascending=False)
train_corr
```

```python
# plotting correlations with coeficient greater than 0.6 (x = dependent variable, y= independent variables)
import seaborn as sns
import matplotlib.style as style

style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (6,4)

for feature in train_corr.index[1:7]:
    sns.regplot(x='SalePrice', y=feature, data=data)
    plt.title(f"{feature}/SalePrice")
    plt.show()
```

Linear correlation with SalePrice is clearly visible, especially with **Gr Liv Area, Total Bsmt SF, Garage Area** and 1st Flr SF features.

```python
# inspecting missing values in train set
train[['Gr Liv Area', 'Total Bsmt SF', 'Garage Area', '1st Flr SF', 'SalePrice']].isnull().sum()
```

```python
# inspecting missing values in test set
test[['Gr Liv Area', 'Total Bsmt SF', 'Garage Area', '1st Flr SF', 'SalePrice']].isnull().sum()
```

```python
# replacing missing values with the appropriate mean values
pd.options.mode.chained_assignment = None
train['Total Bsmt SF'].fillna(train['Total Bsmt SF'].mean(), inplace=True)
test['Garage Area'].fillna(test['Garage Area'].mean(), inplace=True)
```

```python
print(train[['Gr Liv Area', 'Total Bsmt SF', 'Garage Area', '1st Flr SF', 'SalePrice']].isnull().sum())
print(test[['Gr Liv Area', 'Total Bsmt SF', 'Garage Area', '1st Flr SF', 'SalePrice']].isnull().sum())
```

#### Building Simple (Univariate) Linear Regression Model

```python
# finding optimal parameter values
from sklearn import linear_model
lm = linear_model.LinearRegression()
for feature in train_corr.index[1:7]:
    lm.fit(train[[feature]], train['SalePrice'])
    print(f"{feature}: ", lm.coef_)
```

**Gr Liv Area** feature (above ground living area in square feet) has the smallest coeficient, which makes it the most appropriate choice for the univariate model.

```python
# train and test prediction in univariate model
lm.fit(train[['Gr Liv Area']], train['SalePrice'])
train_predictions = lm.predict(train[['Gr Liv Area']])
test_predictions = lm.predict(test[['Gr Liv Area']])

print(f"train predictions: {train_predictions}")
print(f"test predictions: {test_predictions}")
```

```python
# calculating error for univariate model using RMSE (root mean square error)
from sklearn.metrics import mean_squared_error

# train error
train_mse = mean_squared_error(train_predictions, train['SalePrice'])
train_rmse = np.sqrt(train_mse)

# test error
test_mse = mean_squared_error(test_predictions, test['SalePrice'])
test_rmse = np.sqrt(test_mse)

print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
```

```python
# calculating error for univariate model using R-square error
from sklearn.metrics import r2_score
train_r2 = r2_score(train['SalePrice'], train_predictions)
test_r2 = r2_score(test['SalePrice'], test_predictions)

print(f"Train R2: {train_r2}")
print(f"Test R2: {test_r2}")
```

#### Building Multiple (Multivariate) Linear Regression Model

```python
features = ['Gr Liv Area', 'Total Bsmt SF', 'Garage Area', '1st Flr SF']

lm.fit(train[features], train['SalePrice'])

train_predictions = lm.predict(train[features])
test_predictions = lm.predict(test[features])

train_mse = mean_squared_error(train['SalePrice'], train_predictions)
train_rmse = np.sqrt(train_mse)

test_mse = mean_squared_error(test['SalePrice'], test_predictions)
test_rmse = np.sqrt(test_mse)

print(train_rmse)
print(test_rmse)
```

```python
train_r2 = r2_score(train['SalePrice'], train_predictions)
test_r2 = r2_score(test['SalePrice'], test_predictions)
print(train_r2)
print(test_r2)
```

**Both RMSE and R2 have better results in multiple linear regression model. Still, the figures are not good enough (R2 should be at least 0.8 or higher), so further analysis and model building is needed, especially better selection of feature.**

#### Feature Selection

```python
data_2 = pd.read_csv('AmesHousing.txt', delimiter='\t')
train_data = data_2[:1460]
test_data = data_2[1460:]

train_data.head()
```

```python
# extracting only numerical columns (int and float)
numerical_train = train_data.select_dtypes(include=['int', 'float'])

# dropping non-relevant columns (info about various selling dates)
numerical_train = numerical_train.drop(['PID', 'Year Built', 'Year Remod/Add', 'Garage Yr Blt', 'Mo Sold', 'Yr Sold'], axis=1)

numerical_train.head()
```

```python
# inspecting NaN values in columns
null_series = numerical_train.isnull().sum()
null_series
```

```python
# selecting only column names with no missing values
full_cols_series = null_series[null_series == 0]
full_cols_series
```

```python
# extracting columns with no Nan values and calculating correlations
train_subset = train[full_cols_series.index]
sorted_corrs = np.abs(train_subset.corr())['SalePrice'].sort_values(ascending=False)
sorted_corrs
```

#### Correlation Matrix Heatmap

- Correlation Matrix Heatmap helps to better visualize correlations and colinearities among the features

```python
# selecting only correlation above 0.3
strong_corrs = sorted_corrs[sorted_corrs > 0.3]

# filtering train_data
corrmat = train_subset[strong_corrs.index].corr()

# ploting heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corrmat, annot=True)
plt.show()
```

According to the heatmap, strongly related pairs (not including target column correlations) are:
- **Gr Liv Area** (Living Area Above Ground in Square Feet) and **TotRms AbvGrd** (Total Rooms Above the Ground)
- **Garage Cars** (Size of Garage in Car Capacity) and **Garage Area** (Size of Garage in Square Feet)

These four columns reflect very similar information (colinearity).

```python
# dropping TotRms AbvGr and Garage Cars columns (reason: Gr Liv Area and Garage Area columns have continuos values, which may contain more hints about the features)
final_corr_cols = strong_corrs.drop(['TotRms AbvGrd', 'Garage Cars'])
train[final_corr_cols.index].info()
```

```python
# checking for missing values in test set
test[final_corr_cols.index].info()
```

```python
# picking features
features = final_corr_cols.drop('SalePrice').index
features
```

```python
# picking target
target = 'SalePrice'
```

```python
# building linear regression model
lm.fit(train[features], train[target])

train_predictions = lm.predict(train[features])
test_predictions = lm.predict(test[features])

train_mse = mean_squared_error(train[target], train_predictions)
test_mse = mean_squared_error(test[target], test_predictions)

train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)

train_r2 = r2_score(train[target], train_predictions)
test_r2 = r2_score(test[target], test_predictions)

print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
print(f"Train r2: {train_r2}")
print(f"Test r2: {test_r2}")
```

#### Removing Low Variance Features

```python
# normalize all features columns (to get all the values in range between 0 and 1)
train_norm = train[features].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
train_norm
```

```python
# compute the variance of all the columns
train_norm.var().sort_values()
```

```python
# removing the column with lowest variance
features = features.drop('Open Porch SF')

# building linear regression model
lm.fit(train[features], train[target])
train_predictions = lm.predict(train[features])
test_predictions = lm.predict(test[features])
train_mse = mean_squared_error(train[target], train_predictions)
test_mse = mean_squared_error(test[target], test_predictions)
train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)
train_r2 = r2_score(train[target], train_predictions)
test_r2 = r2_score(test[target], test_predictions)

print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
print(f"Train r2: {train_r2}")
print(f"Test r2: {test_r2}")
```

**The figures look better now, but still not good enough. Further feature analysis and more precise selection is needed.**

```python
plt.figure(figsize=(12,8))
sns.kdeplot(test[target], color='b', label='Actual Values')
sns.kdeplot(test_predictions, color='r', label='Predicted Values')
plt.title('Difference Between Actual Values and Predicted Values')
plt.legend()
plt.show()
```

### Modelling with All of the Columns (Non-Numerical Columns Included)

```python
data_copy = data.copy()
data.head()
```

```python
# extracting only non-numerical columns
text_cols = data_copy.select_dtypes(include=['object'])

# checking for missing values
text_cols_mv = text_cols.isnull().sum()
text_cols_mv
```

```python
# only no-null columns
text_cols_nonulls = text_cols_mv[text_cols_mv == 0]
text_cols_nonulls
```

```python
# only non-numerical non-null columns
text_cols_nonulls = text_cols[text_cols_nonulls.index]
text_cols_nonulls.head()
```

```python
# converting non-numerical non-null columns into dummy columns (to get numerical values)
col_dummies = pd.get_dummies(text_cols_nonulls)
col_dummies.head()
```

```python
# delete non-numerical columns from main dataframe
data_copy.drop(text_cols.columns, axis=1, inplace=True)
data_copy.head()
```

```python
# concatenate non-numerical dummies and main dataframe
data_final = pd.concat([data_copy, col_dummies], axis=1)
data_final.head()
```

```python
# checking for missing values
nulls = data_final.isnull().sum()
null_data = nulls[nulls > 0]
null_data
```

```python
# replacing missing values with average values
data_final = data_final.fillna(data_final.mean())
nulls = data_final.isnull().sum()

# confirming there no more missing values
null_data = nulls[nulls > 0]
null_data
```

```python
# data engineering new column
import warnings
warnings.filterwarnings('ignore')
data_final['year_until_remod'] = data_final['Year Remod/Add'] - data_final['Year Built']
data_final = data_final.drop(['Year Remod/Add', 'Year Built'], axis=1)
data_final.head()
```

```python
# checking correlations in new dataframe
corrs = abs(data_final.corr()['SalePrice']).sort_values(ascending=False)
high_corrs = corrs[corrs > 0.5]
high_corrs
```

```python
# plotting correlation matrix heatmap
plt.figure(figsize=(12,8))
sns.heatmap(data_final[high_corrs.index].corr(), annot=True)
plt.show()
```

```python
# only high_corrs columns
data_final = data_final[high_corrs.index]

# remove correlated columns and date columns
data_final.drop(['Garage Cars', 'Total Bsmt SF', 'Garage Yr Blt'], axis=1, inplace=True)
data_final
```

```python
# defining features and target columns
y = data_final['SalePrice']
X = data_final.drop(['SalePrice'], axis=1)

# splitting into train/test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

```python
# building the model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lm = LinearRegression()
lm.fit(X_train, y_train)
y_pred = lm.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print(f"RMSE: {rmse}")
print(f"R2: {r2}")
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(y_test, color='b', label='Actual Values')
sns.kdeplot(y_pred, color='r', label='Predicted Values')
plt.title('Difference Between Actual Values and Predicted Values')
plt.legend()
plt.show()
```

#### Conclusion

The goal of the project was to predict the house sale prices by the use of linear regression. Several different approaches was taken to tackle the problem: simple linear regression and multiple linear regression with different feature selection. The error between actual and predicted values was gradually minimized to an acceptable level (from R2=0.50 to R2=0.82), but further work could be done in feature selection department to get the error figures even lower.

According to this project's results, the properties of a house that most affect the final sale price are as follows (in no particular order):
- Overall material and finish of the house
- Above ground living area (in sqft)
- Size of garage (in sqft)
- The quality of the material on the exterior
- Full bathrooms above ground
- Kitchen quality of excellent and/or average type
- Poured concrete type of foundation
- Masonry veneer area (in sqft)
